Data visualizations for Progress Report 

prelim goal: predict the number of likes a tweet X will recieve

In general our guesses are close, this is just a simple algorithm to guess likes.

```python
import sys, os
import matplotlib.pyplot as plt
import math

# Issues with importing in notebook, so using pickle after running main.py
# Pickling predictions in main, then seeing them here

import pickle

with open('trump.pkl', 'rb') as f:
    trump = pickle.load(f)
with open('avengers.pkl', 'rb') as f:
    avengers = pickle.load(f)
with open('russia.pkl', 'rb') as f:
    russia = pickle.load(f)
with open('top20.pkl', 'rb') as f:
    top_twenty = pickle.load(f)
```

```python
# Graphing our prediction vs actual.
# [text, predicted, like_count]
x,y,x2,y2 = [], [], [], []
X_LABEL = "Tweet number"
Y_LABEL = "Like number"
z = 0
for thing in trump:
    x.append(z)
    y.append(thing[1])
    x2.append(z)
    y2.append(thing[2])
    z+=1

# Our guess is GREEN, actual RED
plt.scatter(x2, y2, c="r")
plt.scatter(x, y, c="g")
plt.xlabel(X_LABEL)
plt.ylabel(Y_LABEL)
plt.show()
print("Our guess is GREEN, actual RED - TRUMP")
# Same chart with reversed layer
plt.scatter(x, y, c="g")
plt.scatter(x2, y2, c="r")
plt.xlabel(X_LABEL)
plt.ylabel(Y_LABEL)
plt.show()
```
